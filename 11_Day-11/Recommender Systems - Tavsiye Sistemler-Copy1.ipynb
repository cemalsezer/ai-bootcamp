```python
### User Based Recommender System
```

```python
import pandas as pd
```

```python
ratings=pd.read_csv('ml-100k/u.data',sep='\t',header=None,names=['user_id','movie_id','rating'],usecols=range(3))
```

```python
ratings.head()
```

```python
movies=pd.read_csv('ml-100k/u.item',sep='|',encoding='iso-8859-1',header=None,names=['movie_id','title'],usecols=range(2))
```

```python
movies
```

```python
ratings=pd.merge(movies,ratings)
```

```python
ratings
```

```python
import numpy as np
```

```python
#Matrix Factorization
```

```python
movieratings=ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
```

```python
movieratings.head()
```

```python
starwars=movieratings['Star Wars (1977)']
```

```python
starwars.head()
```

```python
movieratings[['Star Wars (1977)', '101 Dalmatians (1996)']].corr()
```

```python
movieratings.corrwith(starwars).sort_values(ascending=False)
```
