```python
import pandas as pd
```

```python
ratings=pd.read_csv('ml-100k/u.data',sep='\t',header=None,names=['user_id','movie_id','rating'],usecols=range(3))
```

```python
ratings.head()
```

```python
movies=pd.read_csv('ml-100k/u.item',sep='|',encoding='iso-8859-1',header=None,names=['movie_id','title'],usecols=range(2))
```

```python
movies
```

```python
ratings=pd.merge(movies,ratings)
```

```python
ratings
```

```python
import numpy as np
```

```python
grouped=ratings.groupby('title').agg({'rating':[np.size,np.sum,np.mean]})
```

```python
grouped.head()
```

```python
popular=grouped.sort_values(('rating','mean'),ascending=False)
```

```python
popular.head()
```

```python
toplam=grouped['rating']['sum'].sum()
```

```python
toplam #toplam verilen oy sayısı
```

```python
popular['percentage']=popular['rating','sum'].div(toplam)*100
```

```python
popular.head()
```

```python
popular.sort_values('percentage',ascending=False)
```
