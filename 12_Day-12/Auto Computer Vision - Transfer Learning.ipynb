```python
img=cv2.imread("cars.jpg")
```

```python
plt.imshow(img);
```

```python
box,label,count=cvlib.detect_common_objects(img)
```

```python
output = draw_bbox(img,box,label,count)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(output)
```

```python
def imageDetect(image):
    import pandas as pd
    img=cv2.imread(image)
    plt.imshow(img);    
    box,label,count=cvlib.detect_common_objects(img)   
    output = draw_bbox(img,box,label,count)
    plt.figure(figsize=(10,10))
    plt.imshow(output)
    items=pd.DataFrame(label).value_counts()
    print(items)
    print(f'Bulunan ürün çeşidi: {items.count()}')
```

```python
imageDetect('mutfak.jpg')
```
